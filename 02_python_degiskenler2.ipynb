#Değişkenlerle 4 işlem
+, -, *, ve / işlemleri kolaylıkla yapılabilir.

Sayılar int değerlerle yazılır. Tamsayı anlamına gelir.

```python
sayi1 = int(input("Lütfen birinci notunu giriniz: "))
sayi2 = int(input("Lütfen ikinci notunu giriniz:"))
#Dört işlemi ismine göre değişkenlerde yazalım.
toplam = sayi1 + sayi2
cikarma = sayi1 + sayi2
carpma = sayi1 * sayi2 #çarpmada yıldız kullanılır
bolme = sayi1 / sayi2
print(f"Toplam:{toplam},Çıkarma:{cikarma},Çarpma:{carpma},Bölme:{bolme}")
```
