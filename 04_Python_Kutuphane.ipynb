örneğin;

```python
#Kütüphaneler her zaman projenin en başına eklenir.
import random #Rasgele içerik veya sayı üretmek için kullanılır.

rastgele_sayı = random.randint(1,12) #1-11 arasında rastgele sayı seç
print(rastgele_sayı)
if rastgele_sayı == 1:
  print("Uğur")
elif rastgele_sayı == 2:
  print("Göktuğ")
elif rastgele_sayı == 3:
  print("Azad")
elif rastgele_sayı == 4:
  print("Ebubekir")
elif rastgele_sayı == 5:
  print("Yusuf")
elif rastgele_sayı == 6:
  print("Ali")
elif rastgele_sayı == 7:
  print("Kerem ")
elif rastgele_sayı == 8:
  print("Ali Kerem")
elif rastgele_sayı == 9:
  print("Kaya")
elif rastgele_sayı == 10:
  print("Haşmet dayı")
elif rastgele_sayı == 11:
  print("Muhammed")
elif rastgele_sayı == 12:
  print("Beril")
```

#Kütüphanelerde kısaltma
Bazen kütüphanelerin ismi uzundur. Biz o isimleri kısaltmak için
**as** takısı kullanarak kütüphane isimlerini kısaltırız.

```python
import random as rd #Kısaltma yapılabiliyor.
import math as m

sayi1 = rd.randint(1,10) #1-10 arasında rastgele sayıyı değişkeni atayabiliyoruz.

#matematiksel fonksiyonlar
print(m.factorial(sayi1)) #faktörel
print(m.comb(1,2)) #kombinasyonun (iki değer atar) (sayi1, sayi2)
print(perm(sayi1, 2)) #pernitasyon
print(m.sqrt(sayi1)) #karekök alma
print(m.floor(sayi1)) #Yuvarlama
print(m.pow(sayi1, 4)) #istediğin sayının kökkuvvetini alma
print(m.log(sayi1,3)) #istediğin tabandan logaritma
print(m.pi) #pi sayısı
print(m.e) # e sayısı
```
