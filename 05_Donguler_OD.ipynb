Toplam iki farklı döngü vardır. Biri **for** diğeri ise **while** döngüsüdür. Şimdi **for döngüsünü** inceleyelim.

```python
#For döngüsünü kurmak için yapılması gerekenler basittir:
for i in range(1,100):
  print(1)
#Her zaman son satırı almaz. O yüzden bir rakam fazla yazarım.
#100 değil 101 yazarım
#yukarıdaki döngüyü inceleyelim:
#for... in ... : bir kalıptır
#i yazan şey bir değişkendir
#range ise bir aralık anlamına gelmektedir.
#Range yerine başka şeyler yazılabilir.
#Bir başka örnek daha yapalım:
ismim = "Kerem Parpar"
for a in ismim:
  print(a)
#Adımın harflerini for döngüsü ile parçalayabilirim.
#Satır satır yazdırabilirim.
```

##For  döngüsünde işlemler

For döngüsünde işlemler sadece sayıları ard arda yazdırmayız. Aynı zamanda bir
çok farklı işlemler yapabiliriz.

```python
ismim = "Kerem Mert Enes"
for degisken in range(0,len(ismim)):
  #print(degisken)
  print(ismim)
```

```python
sayi1=int(input("Lütfen döngüye başlayacağı sayıyı yazınız:"))
sayi2=int(input("Lütfen döngüye biteceği sayıyı yazınız:"))
for degisken in range(sayi1,sayi2):
  print(degisken)
```

```python
for a in range(0,20,2):
  print(a)
```

```python
for a in range(20,0,-2):
  print(a)
```

**For döngüsünde range() üç değer alabilir. Birincisi başlangıç, ikincisi bitiş
değeri, üçüncü ise döngünün kaçar kaçar artacağını gösterir.**

```python
for a in range(0,50):
  if (a % 5 ==0):
    print(a)
  else:
    print("Bu sayı o sayı değil.")
```

**NOT**: **% (Yüzde)** işareti sayının modunu alma anlamına gelir.
Örneğin bir sayının bölümünden kalanı verir. Mesela "a" sayısını 6'ya bölünümünden kalanına **mod** denir.

```python
for i in range(0,50):
  if(i % 5 == 0 or i % 3 == 0 ):
    print("15'e tam bölünen sayılar:", i)
  if(i % 5 == 0 or i % 3 == 0):
    print(i)
```

**YAPILACAK:** Bilgisayar rastgele iki sayı üretilecek ve üretilen döngü kurulacak.

Birinci sayı: 0-50 arasında olacak

İkinci sayı:60-250 arasında olacak

**Döngü 3'er 3'er artacak**

```python
import random
rastgele_sayi1 = random.randint(0,50)
rastgele_sayi2 = random.randint(60,250)
for i in range(rastgele_sayi1,rastgele_sayi2,3):
  print(i)
```

#iç içe For Döngüleri olabilir
Bazen birden fazla döngüsü birbirinin içinde olabilir.Bu biraz karmaşıktır. Çok fazla pratik yapmak gerekir.

```python
for a in range(0,5):
  print("a değişkeni şunu yazdı:", a)
  for b in range(0,5):
    print("b değişkeni bunu yazdı:", b)
```

```python
#ödev
sayi1 = int(input("Lütfen ilk sayıyı giriniz:"))
sayi2 = int(input("Lütfen ikinci sayıyı giriniz:"))
for i in range(sayi1,sayi2):
  if(i % 2 == 0 or i % 3 == 0):
    print(i)
  else:
    print("Bu sayı o sayı değil")
```

```python
for i in range(0, 51):
    if i % 15 == 0:
        print("15'e bölünen sayılar: ", i)
    if i % 3 == 0:
        print("3'e bölünen sayılar: ", i)
```

```python
for a in range(1,100):
  for b in range(1,100):
    print(a,"içinde dönen b döngüsü değeri:",b)
```

```python
#For döngüsünün içindeki sayıları toplama
toplam = 0
bilgi = int(input("Lütfen 1'den itibaren kaça kadar toplayacağının sayıyı yazınız:"))
for sayi in range(1,bilgi+1):
  toplam = toplam + sayi
  print("Toplam değişkeninin değeri:",toplam,"Eklenen sayı ise:",sayi)
print("Toplam:",toplam)
```
