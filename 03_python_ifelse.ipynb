#Python'da IF-Else Koşulları
Python'da değişkenlerin içindeki değerlere göre işlem
yapabilirim. Programlarımızda kullanıcıdan aldığımız değerlere
göre koşullar oluşturup kullanıcının isteğine göre işlemler
yapabilirim.

```python
#IF-Else kullanımı
#Boolean değişkeni sadece true(doğru) veya false(yanlış) değeri alan değişkenlerdir.
cinsiyet = True
if cinsiyet == True: #Eğer cinsiyet kadınsa
  print("Merhaba hanımefendi")
else: #değilse (kadın değilse)
  #İf veya else içinde tab ile boşluk bırakarak yazarım.
  print("Merhaba beyfendi")
print("Program bitti...")
```
