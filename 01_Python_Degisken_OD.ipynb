#DEĞİŞKENLER
Verileri saklamamızı sağlayan yapılardır. Diyelim ki kullanıcı bilgisayara bilgi girdi. Girilen bilgiyi ben programımda işlemek istersem onu bir değişkenin içinde saklarım.

```python
#Bir değişken oluşturalım.
#Değişken adimi ve soyadimi saklasin
ad_soyad = "Kerem Parpar"
yas = 16
print(ad_soyad) #Değişkenler print edilmezse ekranda gözükmez!
print(yas)
```

#Değişkenlerin özellikleri

1. Değişkenler amacına göre isimlendirilir.
2. İsim verilirken Türkçe karakter kullanılmaz
3. Değişkenler, sayı, metin, kesirli sayı, Doğru/Yanlış değerlerini alabilir.

```python
#Farklı değişken türleri
arabanin_ismi = "Bmw" #Metin -> String  (str) demektir.
arabanin_motor_gucu = 500 #Sayı -> Integer (int) demektir
arabanin_yuksekligi = 50 #Yükseklik -> Float (f) ""
arabanin_genisligi = 95 #Ondalık sayı -> Double (d) ""
araba_calisiyor_mu = True #Doğru/Yanlış -> Boolean (bool) ""
araba_pahali_mi = True #Doğru/Yanlış -> Boolean (bool) ""
print(arabanin_ismi, arabanin_motor_gucu, arabanin_yuksekligi, arabanin_genisligi, araba_calisiyor_mu, araba_pahali_mi)
```
